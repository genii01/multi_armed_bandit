```python
#import library
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
```

```python
class BernoulliTSAgent(object):
    
    def __init__(self):
        '''beta distribution initialization 
        (alpha, beta => 1,1 uniform distribution setup)'''
        
        self.counts = [0 for _ in range(n_arms)]
        self.wins = [0 for _ in range(n_arms)]
        
    def get_arm(self):
        '''arm select distribution and sampling'''
        
        # beta distribution function 생성 (sampling # 1,000)
        beta = lambda N, a : np.random.beta(a+1, N-a+1,1000)
        
        # arm select distribution (beta distribution 기반)
        arms_distribution = []
        
        for i in range(n_arms):
            arm_select_dist = beta(self.counts[i], self.wins[i])
            arms_distribution.append(arm_select_dist)
        return arms_distribution
```

```python
n_arms = 10
agent = BernoulliTSAgent()
print(agent.counts)
print(agent.wins)
res = agent.get_arm()
print(res)
```

```python
# 최종 multi_armed_bandit 적용 클래스
class BernoulliTSAgent(object):
    
    def __init__(self):
        self.counts = [0 for _ in range(n_arms)]
        self.wins = [0 for _ in range(n_arms)]
        
    def get_arm(self):
        beta = lambda N, a : np.random.beta(a+1, N-a+1)
        result = [beta(self.counts[i], self.wins[i]) for i in range(n_arms)]
        return result
    
```

```python
agent = BernoulliTSAgent()
agent.counts = [101,155,156,1500,141,133,134,134,191,1000]
agent.wins = [9,12,17,876,21,17,16,11,5,511]
res = agent.get_arm()
print(res)
```
